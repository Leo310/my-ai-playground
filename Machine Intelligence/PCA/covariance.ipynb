```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sd = 10
```

```python
mean = (0, 0)
cov1 = [[1, 0],
       [0, 1]]
cov2 = [[1, 0],
       [0, 2]]
cov3 = [[1, -1],
       [-1, 1]]
cov4 = [[1, -0.5],
       [-0.5, 1]]
covs = [cov1, cov2, cov3, cov4]
cmaps = ["Blues_r", "Reds_r", "Greens_r", "Purples_r"]
fig, axs = plt.subplots(2,2,figsize=(10,10), sharex=True, sharey=True)
print(axs)
k=0
np.random.seed(sd);
for idx, cov in enumerate(covs):
    data = np.random.multivariate_normal(mean, cov, 500)
    # plt.scatter(data[:,0], data[:,1])
    r = k//2
    c = k%2
    plt.sca(axs[r][c])
    sns.kdeplot(x=data[:,0], y=data[:,1], bw_method=.99, cmap=cmaps[idx])#, shade=False, bw=.15)
    if r > 0:
        plt.xlabel(r'$x_1$', fontsize=20)
    if c == 0: 
        plt.ylabel(r'$x_2$', fontsize=20, rotation='horizontal')
    plt.title('ABCD'[k], fontsize=25)
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.xlim([-max(np.abs(plt.axis())), max(np.abs(plt.axis()))])
    plt.ylim([-max(np.abs(plt.axis())), max(np.abs(plt.axis()))])
    plt.grid()
    k+=1
plt.tight_layout()
fig.savefig('cov.pdf')
```
